# 使用PCA对图像进行降维，k=36

```python
import numpy as np
import scipy.io as sio
import matplotlib.pyplot as plt
```

```python
mat = sio.loadmat('ex7faces.mat')
X = mat['X']
X.shape
```

```python
def plot_100_images(X):
    fig,axs = plt.subplots(ncols=10,nrows=10,figsize=(10,10))
    for c in range(10):
        for r in range(10):
            axs[c,r].imshow(X[10*c+r].reshape(32,32).T , cmap = 'Greys_r')#显示单通道的灰度图
            axs[c,r].set_xticks([])
            axs[c,r].set_yticks([])
```

```python
plot_100_images(X)
```

```python
X_demean = X - np.mean(X,axis = 0)
C = X_demean.T @ X_demean / len(X)
U,S,V = np.linalg.svd(C)
U1 = U[:,:36]
X_reduction = X_demean @ U1
X_reduction.shape
```

```python
X_recover  = X_reduction@U1.T + np.mean(X)
```

```python
plot_100_images(X_recover),plot_100_images(X)
```
